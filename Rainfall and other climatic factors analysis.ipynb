```python
import pandas as pd
import matplotlib.pyplot as plt

# Converting 'time' to datetime format
df['time'] = pd.to_datetime(df['time'], errors='coerce')

# Filtering data by location (sampling every 10th row)
bengaluru_data = df[df['Location'] == 'Bengaluru'].iloc[::10]
delhi_data = df[df['Location'] == 'Delhi'].iloc[::10]
chennai_data = df[df['Location'] == 'Chennai'].iloc[::10]
kolkata_data = df[df['Location'] == 'Kolkata'].iloc[::10]

# Scatter plot for Precipitation (mm)
plt.figure(figsize=(10, 6))
plt.scatter(bengaluru_data['time'], bengaluru_data['precipitation(mm)'], label='Bengaluru', color='blue', alpha=0.6)
plt.scatter(delhi_data['time'], delhi_data['precipitation(mm)'], label='Delhi', color='red', alpha=0.6)
plt.scatter(chennai_data['time'], chennai_data['precipitation(mm)'], label='Chennai', color='green', alpha=0.6)
plt.scatter(kolkata_data['time'], kolkata_data['precipitation(mm)'], label='Kolkata', color='purple', alpha=0.6)

plt.xlabel('Time')
plt.ylabel('Precipitation (mm)')
plt.title('Precipitation Over Time for Different Locations')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter plot for Temperature at 80m (°C)
plt.figure(figsize=(10, 6))
plt.scatter(bengaluru_data['time'], bengaluru_data['temperature_80m (°C)'], label='Bengaluru', color='blue', alpha=0.6)
plt.scatter(delhi_data['time'], delhi_data['temperature_80m (°C)'], label='Delhi', color='red', alpha=0.6)
plt.scatter(chennai_data['time'], chennai_data['temperature_80m (°C)'], label='Chennai', color='green', alpha=0.6)
plt.scatter(kolkata_data['time'], kolkata_data['temperature_80m (°C)'], label='Kolkata', color='purple', alpha=0.6)

plt.xlabel('Time')
plt.ylabel('Temperature at 80m (°C)')
plt.title('Temperature at 80m Over Time for Different Locations')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter plot for Rain (mm)
plt.figure(figsize=(10, 6))
plt.scatter(bengaluru_data['time'], bengaluru_data['rain(mm)'], label='Bengaluru', color='blue', alpha=0.6)
plt.scatter(delhi_data['time'], delhi_data['rain(mm)'], label='Delhi', color='red', alpha=0.6)
plt.scatter(chennai_data['time'], chennai_data['rain(mm)'], label='Chennai', color='green', alpha=0.6)
plt.scatter(kolkata_data['time'], kolkata_data['rain(mm)'], label='Kolkata', color='purple', alpha=0.6)

plt.xlabel('Time')
plt.ylabel('Rain (mm)')
plt.title('Rain Over Time for Different Locations')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter plot for Rain (mm)
plt.figure(figsize=(10, 6))
plt.scatter(bengaluru_data['time'], bengaluru_data['soil_temp(°C)'], label='Bengaluru', color='blue', alpha=0.6)
plt.scatter(delhi_data['time'], delhi_data['soil_temp(°C)'], label='Delhi', color='red', alpha=0.6)
plt.scatter(chennai_data['time'], chennai_data['soil_temp(°C)'], label='Chennai', color='green', alpha=0.6)
plt.scatter(kolkata_data['time'], kolkata_data['soil_temp(°C)'], label='Kolkata', color='purple', alpha=0.6)

plt.xlabel('Time')
plt.ylabel('Soil temperature')
plt.title('Soil temperature over the period')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

```python
import seaborn as sns
import numpy as np


# Ensuring that 'time' is in datetime format and adding a 'day' column for daily aggregation
df['time'] = pd.to_datetime(df['time'], errors='coerce')
df['day'] = df['time'].dt.date  # Extract date for daily aggregation

# Aggregate data (daily mean for each location)
daily_data = df.groupby(['day', 'Location']).mean().reset_index()

# Pivot data to have locations as columns for each metric (e.g., temperature, precipitation)
pivot_temp = daily_data.pivot(index='day', columns='Location', values='temperature_80m (°C)')
pivot_precip = daily_data.pivot(index='day', columns='Location', values='precipitation(mm)')
pivot_rain = daily_data.pivot(index='day', columns='Location', values='rain(mm)')
pivot_soil_temp = daily_data.pivot(index='day', columns='Location', values='soil_temp(°C)')


# Clustermap for temperature at 80m across locations over days
plt.figure(figsize=(10, 8))
sns.clustermap(pivot_temp, cmap="coolwarm", figsize=(12, 8), metric='euclidean', method='ward')
plt.title('Clustered Heatmap of Temperature at 80m Across Days and Locations')
plt.show()

# Clustermap for precipitation across locations over days
plt.figure(figsize=(10, 8))
sns.clustermap(pivot_precip, cmap="Blues", figsize=(12, 8), metric='euclidean', method='ward')
plt.title('Clustered Heatmap of Precipitation Across Days and Locations')
plt.show()

# Heatmap for rain across locations over days (non-clustered for simple time-based analysis)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_rain, cmap="YlGnBu", cbar_kws={'label': 'Rain (mm)'})
plt.title('Heatmap of Rain Across Days and Locations')
plt.xlabel('Location')
plt.ylabel('Day')
plt.show()

# Heatmap for soil temperature across locations over days
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_soil_temp, cmap="Oranges", cbar_kws={'label': 'Soil Temperature (°C)'})
plt.title('Heatmap of Soil Temperature Across Days and Locations')
plt.xlabel('Location')
plt.ylabel('Day')
plt.show()
```
